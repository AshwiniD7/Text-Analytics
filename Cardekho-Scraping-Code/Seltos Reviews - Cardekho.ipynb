```python
#Import Libraries required for web scraping
from selenium import webdriver
from selenium.webdriver.support import expected_conditions as EC
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
import time
```

```python
#This code will open the chrome browser and then open the cardekho website
#To run this code, Provide your chromedriver.exe path
driver = webdriver.Chrome(ChromeDriverManager().install())
driver.get('https://www.cardekho.com/kia/seltos/user-reviews')
```

```python
#This code will find the url link of the 'full review' on the first page. 
url=[]
for data in driver.find_elements_by_class_name('readReviewHolder'):
    url.append(data.find_element_by_tag_name('a').get_attribute('href'))
```

```python
#This code will find the url link of the 'full review' on all the 20 pages.
for i in range(2,21,1):
    cook_url = 'https://www.cardekho.com/kia/seltos/user-reviews/'+ str(i) +'?subtab=latest'
    driver.get(cook_url)
    for data in driver.find_elements_by_class_name('readReviewHolder'):
        url.append(data.find_element_by_tag_name('a').get_attribute('href'))
```

```python
#Print the number of customer reviews
len(url)
```

```python
#Print all the customer reviews URLs
url
```

```python
#This code will fetch review information from all the review URLs. 
car_review=[]
brand = 'Kia Seltos'
for each_url in url:
    print(each_url)
    driver.get(each_url)
    time.sleep(1)
    data = driver.find_element_by_class_name('readReviewHolder')
    review = data.find_element_by_class_name('contentheight').text
    print(review)
    author_name = data.find_element_by_class_name('name').text[3:]
    print(author_name)
    review_date = data.find_element_by_class_name('date').text[4:16]
    print(review_date)
    rating=0
    try:
        for i in range(1,6,1):
            data.find_element_by_class_name('stars.stars'+str(i)+'.icon-star-full-fill ')
            rating+=1
    except EC.NoSuchElementException:
        pass
    print(rating)
    car_review.append({'Brand':brand,'Review':review,'Rating':rating,'Author Name': author_name, 'Review Date': review_date,'URL':each_url})
```

```python
#Save customer review in csv format
final_data = pd.DataFrame(car_review, columns=['Brand','Review','Rating','Author Name','Review Date','URL'])
final_data.to_csv('kiaseltos_reviews.csv')
```
